# Signature diagnosis

Once a signature has been created, it can be diagnosed using multiple plotting functions.

Let's analyze, for instance, the **signature 2** of **B1.001** (MoA)

## Set up

We start by instantiating a Chemical Checker class.

```python
from chemicalchecker import ChemicalChecker
cc = ChemicalChecker()
```

Our signature of interest can be loaded as follows:

```python
sign = cc.signature("B1.001", "sign2")
```

Now we can instantiate a diagnosis class focused on this signature:

```python
diag = cc.diagnosis(sign)
```

## Individual plots

In order to obtain plots, data needs to be precomputed. By default, the methods will look in the `stats` folder corresponding to the signature. If the needed file is not available, it will be produced on the fly.

The following gives the list of available diagnosis plots:

```python
diag.available()
```

Let's see, for example, the 2D projection of our signature:

```python
%matplotlib inline
ax = diag.projection()
```

## Canvas

Finally, an overview of the diagnosis can be obtained in a canvas:

```python
diag.canvas()
```
